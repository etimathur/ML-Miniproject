```
# K Means Clustering Algorithm
# Dataset consists of 2 columns population and the number of people vaccinated in that particular locality
# We classify them into two clusters based on the similarities in the parameters
# Importing modules
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```
# Importing dataset
dataset = pd.read_csv('vaccine.csv')
X = dataset.iloc[:, [0,1]].values
print(X)
```

```
# Finding the appropraite number of clusters according to the dataset
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```
# Using K Means Algorithm with no of clusters=2
kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

```
# Plotting the clusters on a graph
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('K means')
plt.xlabel('Population')
plt.ylabel('No of people vacinated')
plt.legend()
plt.show()
```
