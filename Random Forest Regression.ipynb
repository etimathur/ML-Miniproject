```python
# Random Forest Regression Algorithm
# Dataset consists of the day and the amount of vaccines distributed in a particular locality
# We use this algorithm to predict the future number of vaccines to be distributed according to the previous trends
# Importing modules
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
data = pd.read_csv('vaccinationtrends.csv')
print(data)
```

```python
# Splitting the dataset
x = data.iloc[:, 0:1].values
print(x)
y = data.iloc[:, 1].values
 
```

```python
print(y)
```

```python
# Using Random Forest Regressor Algorithm
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)

regressor.fit(x, y) 
```

```python
# Predicting for test data
Y_pred = int(regressor.predict(np.array([51]).reshape(1, 1)))
print(Y_pred)
```

```python
# Plotting the graph
X_grid = np.arange(min(x), max(x), 0.01)               
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(x, y, color = 'blue')  
plt.plot(X_grid, regressor.predict(X_grid), color = 'green') 
plt.title('Random Forest Regression')
plt.xlabel('Day')
plt.ylabel('Vaccines Distributed')
plt.show()
```
